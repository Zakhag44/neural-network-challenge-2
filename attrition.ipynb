## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python

# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition','Department']]
```

```python
# Create a list of at least 10 column names to use as X data

x_column_list = ['Age', 'BusinessTravel', 'Department', 'TotalWorkingYears', 'Education',
                 'EducationField', 'YearsSinceLastPromotion', 'DistanceFromHome', 'EnvironmentSatisfaction',
                 'TrainingTimesLastYear', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction',
                 'MaritalStatus', 'PerformanceRating', 'YearsAtCompany', 'NumCompaniesWorked',
                 'YearsWithCurrManager', 'OverTime']


# Create X_df using your selected columns
X_df = attrition_df[x_column_list]

# Summarize the data types for X_df
X_df.info()


```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2, random_state=42)

```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# Ensure both training and testing data have the same columns after encoding
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit_transform(X_train)

# Scale the training and testing data
X_test_scaled = scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
encoder_dep = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
y_train_dep = encoder_dep.fit_transform(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
y_test_dep = encoder_dep.transform(y_test[['Department']])

```

```python
# Create a OneHotEncoder for the Attrition column
encoder_attr = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
y_train_attr = encoder_attr.fit_transform(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_test_attr = encoder_attr.transform(y_test[['Attrition']])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_shape = X_train_scaled.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(input_shape,))

# Create at least two shared layers
shared_layer_1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer_2 = layers.Dense(32, activation='relu')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

#hidden layer
dept_hidden_layer = layers.Dense(16, activation='relu')(shared_layer_2)


# Create the output layer
dept_output = layers.Dense(y_train_dep.shape[1], activation='softmax', name='department_output')(dept_hidden_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attr_hidden_layer = layers.Dense(16, activation='relu')(shared_layer_2)

# Create the output layer
attr_output = layers.Dense(y_train_attr.shape[1], activation='softmax', name='attrition_output')(attr_hidden_layer)

```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output, attr_output])

# Compile the model
model.compile(optimizer='adam',
                  loss={'department_output': 'categorical_crossentropy',
                        'attrition_output': 'categorical_crossentropy'},
                  metrics={'department_output': 'accuracy',
                           'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
history = model.fit(X_train_scaled, {'department_output': y_train_dep, 'attrition_output': y_train_attr},
                        epochs=100, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model with the testing data
model.evaluate(X_test_scaled, {'department_output': y_test_dep, 'attrition_output': y_test_attr})
```

```python
# Print the accuracy for both department and attrition
dept_acc = history.history['department_output_accuracy']
attr_acc = history.history['attrition_output_accuracy']
print(f"Department predictions accuracy: {dept_acc[-1]}")
print(f"Attrition predictions accuracy: {attr_acc[-1]}")
```
